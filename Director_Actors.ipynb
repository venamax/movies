```python
df_plt = df_plot[['plot']]
```

```python
df_set = df_plt.join(df)
```

```python
df_set['revenues']=pd.to_numeric(df_set['revenues'], errors='coerce')
```

```python
df_set['costs']=pd.to_numeric(df_set['costs'], errors='coerce')
```

```python
df_set.count()
```

```python
df_set = df_set.reset_index().drop_duplicates(subset='index', keep='last').set_index('index')
```

```python
df_set.count()
```

```python
directors = {}
for index, row in df.iterrows():
    director = row[u'Directed by']
    try:
        if director not in directors:
            directors[director] = []
        directors[director].append(row[u'site'])
    except ValueError:
        continue

            
```

```python
movie_directors = {}
for index, row in df.iterrows():
    director = row[u'Directed by']
```

```python
df_dir = pd.DataFrame(directors.items())
```

```python
df_dir.index=df_dir[0]
```

```python
df_dir.columns = ['director_name', 'movies']
```

```python
df_dir
```

```python
dir_revenues=[]
dir_costs=[]
for index, row in df_dir.iterrows():
    movies = row['movies']
    sum_box_office=0
    sum_budget = 0
    for movie in movies:
        try:
            if (df_set['revenues'][movie]>0) & (df_set['costs'][movie]>0): 
                sum_box_office = sum_box_office + df_set['revenues'][movie]
                sum_budget = sum_budget + df_set['costs'][movie]
        except ValueError:
            continue
    dir_revenues.append(sum_box_office)
    dir_costs.append(sum_budget)
        
```

```python
df_dir['revenues']=dir_revenues
df_dir['costs']=dir_costs
```

```python
df_dir['gross_profit']= df_dir['revenues'].apply(lambda x : x*0.5) - df_dir['costs'].apply(lambda y : y)
```

```python
df_dir['return']=df_dir['gross_profit']/df_dir['costs']
```

```python
df_dir['count']=df_dir['movies'].apply(lambda x : len(x))
```
